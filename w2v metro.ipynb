```python
import bs4 as bs
import urllib.request
import re
import nltk

urllist= ['https://www.mumbai77.com/city/3315/travel/metro-trains/',  'http://beta.reliancemumbaimetro.com/about-us.aspx', 'http://beta.reliancemumbaimetro.com/metro-train-schedule.aspx', 'http://beta.reliancemumbaimetro.com/how-to-use.aspx#metro-entrances-and-exits', 'http://beta.reliancemumbaimetro.com/how-to-use.aspx#where-to-buy-tickets',
'http://beta.reliancemumbaimetro.com/how-to-use.aspx#use-of-elevators-and-escalators', 'http://beta.reliancemumbaimetro.com/how-to-use.aspx#afc-gates', 
'http://beta.reliancemumbaimetro.com/how-to-use.aspx#reserved-seats',
'http://beta.reliancemumbaimetro.com/metro-faq.aspx#Fare', 'http://beta.reliancemumbaimetro.com/metro-faq.aspx#TravelTrain', 'http://beta.reliancemumbaimetro.com/metro-faq.aspx#Station', 'http://beta.reliancemumbaimetro.com/metro-faq.aspx#TravelTrain', 
'http://beta.reliancemumbaimetro.com/metro-faq.aspx#Project',
'http://beta.reliancemumbaimetro.com/metro-faq.aspx#SmartCard', 'http://beta.reliancemumbaimetro.com/metro-faq.aspx#Token', 'http://beta.reliancemumbaimetro.com/metro-faq.aspx#TravelTrain', 'http://beta.reliancemumbaimetro.com/metro-faq.aspx#Station',
'http://beta.reliancemumbaimetro.com/metro-faq.aspx#security-safety-emergency', 'http://beta.reliancemumbaimetro.com/metro-faq.aspx#Others',
'http://beta.reliancemumbaimetro.com/metro-customer-care.aspx',
'http://beta.reliancemumbaimetro.com/security.aspx#know-your-security', 'http://beta.reliancemumbaimetro.com/security.aspx#faq',
'http://beta.reliancemumbaimetro.com/features.aspx#StationFacilities', 
'http://beta.reliancemumbaimetro.com/features.aspx#TrainFeatures',
'http://beta.reliancemumbaimetro.com/working-of-metro-train.aspx',
'http://beta.reliancemumbaimetro.com/conditions-of-carriage.aspx',
'http://beta.reliancemumbaimetro.com/my-metro-my-responsibility.aspx#ticketing',
'http://beta.reliancemumbaimetro.com/my-metro-my-responsibility.aspx#baggage-boarding-and-alighting',
'http://beta.reliancemumbaimetro.com/my-metro-my-responsibility.aspx#responsible-citizens',
'http://beta.reliancemumbaimetro.com/offences-penalties.aspx']


for x in urllist:
    scrapped_data = urllib.request.urlopen(x)
    article = scrapped_data. read()
    with open("copy.txt", "a") as file:
        file.write("\n")
        file.write(str(article))
    
        
temp_file= open('copy.txt')
parsed_article = bs.BeautifulSoup(temp_file,'lxml')

paragraphs = parsed_article.find_all('p')

article_text = ""

for p in paragraphs:
    article_text += p.text
```

```python
# Cleaning the text
processed_article = article_text.lower()
processed_article = re.sub('[^a-zA-Z]', ' ', processed_article )
processed_article = re.sub(r'\s+', ' ', processed_article)

# Preparing the dataset
all_sentences = nltk.sent_tokenize(processed_article)

all_words = [nltk.word_tokenize(sent) for sent in all_sentences]

# Removing Stop Words
from nltk.corpus import stopwords
for i in range(len(all_words)):
    all_words[i] = [w for w in all_words[i] if w not in stopwords.words('english')]
```

```python
from gensim import models
word2vec = models.Word2Vec(all_words, min_count=1)
```
